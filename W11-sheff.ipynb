```python
# Compute features: mean, variance, min, max, count
from collections import Counter

# Create new fields
for field in ['mean', 'var', 'count', 'min', 'max']:
  results[field] = 0

def get_features(reviews):
  '''Takes in a list of movie ratings and returns a dictionary with `mean`, `var`, `min`, `max` and `count`
  '''
  reviews = list(filter(lambda x: x >= 0, reviews)) # list of reviews in above cell
  
  print(Counter(reviews).keys())
  print(Counter(reviews).values())
  
  features_list = {                      # list of size 5 containing the below items
    'mean': np.mean(reviews),
    'var': np.var(reviews),
    'min': min(reviews),
    'max': max(reviews),
    'count': len(reviews)
    # count_1
    # count_2
    # count_3
    # count_4
    # count_5
  }

  return features_list
```
